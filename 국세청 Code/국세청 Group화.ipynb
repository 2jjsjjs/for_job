원래는 경상소득(조사) 가 경상소득을 모두 합친거이기 때문에 다른 경상소득을 이루는 하위소득들보다 작으면 안된다.

```python
new_df.loc[new_df['경상소득(조사)'] < new_df['경상소득_근로소득(조사)']]
```

여기서 CTGAN의 Constraint 조건이 들어간다.

```python
from sdv.constraints import Inequality
```

```python
inc_constraint = Inequality(
    low_column_name='경상소득_근로소득(조사)',
    high_column_name='경상소득(조사)'
)
```

```python
constraints = [inc_constraint]
```

```python
start = time.time()
model = CTGAN(primary_key='MD제공용_가구고유번호' , epochs = 30 , batch_size = 400 , verbose = True ,constraints = constraints)
model.fit(oridata)
model.save('C:/Users/2jjsj/test_model_fill_outlier_fin.pkl')

end = time.time()
print(f"{end - start:.5f} sec")  
```

#### 층별로 ctgan 구현하기 

```python
sex = [oridata['가구주_성별코드'] == 1 , oridata['가구주_성별코드'] == 2]
condition = [oridata['경상소득(조사)'] < 1800 , 1800 < = oridata['경상소득(조사)'] < 3704]
```

```python
, 1800 < = oridata['경상소득(조사)'] < 3704 , 3704 < = oridata['경상소득(조사)'] < 6300 , 6300 < = oridata['경상소득(조사)']
```

```python
sex
```

```python
import numpy as np

oridata['ind_c'] = np.where(oridata["경상소득(조사)"] < 1800, 1,
                            np.where(oridata["경상소득(조사)"] < 3704, 2,
                                     np.where(oridata["경상소득(조사)"] < 6300, 3,
                                              np.where(oridata["경상소득(조사)"] >= 6300, 4, None))))
```

```python
index = np.array([(x, y) for x in range(1,3) for y in range(1,5)])
index1 = pd.DataFrame(index)
index1
```

```python
data_list = [] # 빈 리스트 생성
 
for i in range(0,8):
    line = []             # 안쪽 리스트로 사용할 빈 리스트 생성
    for j in range(0,8):
        line.append(oridata.loc[(oridata['가구주_성별코드'] == index1.iloc[j,0]) & (oridata['ind_c'] == index1.iloc[j,1]),]) 
    data_list.extend(line)         # 전체 리스트에 안쪽 리스트를 추가
 
data_list
```

```python
start = time.time()

new_sample = []

model1 = CTGAN(primary_key='MD제공용_가구고유번호' , epochs = 30 , batch_size = 400 , verbose = True)
model2 = CTGAN(primary_key='MD제공용_가구고유번호' , epochs = 40 , batch_size = 450 , verbose = True)
model3 = CTGAN(primary_key='MD제공용_가구고유번호' , epochs = 35 , batch_size = 500 , verbose = True)

for i in range(0,8):
    if len(data_list[i]) < 3000 : 
        model1.fit(data_list[i])
        new_sample.append(model1.sample(len(data_list[i])))          
        
    elif 3000 <= len(data_list[i]) < 5000 : 
        model2.fit(data_list[i])
        new_sample.append(model2.sample(len(data_list[i])))           
        
    else : 
        model3.fit(data_list[i])
        new_sample.append(model3.sample(len(data_list[i])))
        
end = time.time()
print(f"{end - start:.5f} sec")       
```
